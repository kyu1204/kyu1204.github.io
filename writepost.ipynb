```python
import sys
import time
```

```python
localtime=time.localtime(time.time())
localtime
year = str(localtime.tm_year)
mon = str(localtime.tm_mon)
day = str(localtime.tm_mday)
fileday = year+'-'+mon+'-'+day+'-'
```

```python
option = sys.argv[1]
mdformat = ['published','title','series','categories','tags','layout','excerpt','comments','toc']

if option == '-c':
    filename = fileday+sys.argv[2]
    with open('_posts/'+filename, 'rw')as f:
        f.write('---')
        for key in mdformat:
            mdf = key+':'+''
            f.write(mdf)
        f.write('---')
```
